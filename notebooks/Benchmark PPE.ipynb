## Using YOLO with pre trained model Benchmark
Will serve as base for analysis

Pre trained model downloaded from:  [Download ppe.pt](https://drive.google.com/drive/folders/11tfTBkp4JdlJ8QXoAMZxgVMf8xpLBXi_?usp=sharing)

# Load image and libs

```python
from ultralytics import YOLO
import cv2
import math
import matplotlib.pyplot as plt

# Support methods for G4
import msaai


# Load image just for show
man_file = "../datasets/man-hard-hat.jpg"
group_file = "../datasets/source_files/schupan_japan1web.jpg"

# Open image file for reading and fix color grading
img_man = cv2.imread(man_file) 
img_man = cv2.cvtColor(img_man, cv2.COLOR_BGR2RGB)

img_group = cv2.imread(group_file) 
img_group = cv2.cvtColor(img_group, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(img_man)
plt.show()
plt.imshow(img_group)
plt.show()
```

# Configure Model & Prediction

Pre trained model:  [Download ppe.pt](https://drive.google.com/drive/folders/11tfTBkp4JdlJ8QXoAMZxgVMf8xpLBXi_?usp=sharing)

```python

# Load YOLO model for detecting objects related to personal protective equipment (PPE)
model = YOLO("../datasets/ppe.pt")

# Class names for different objects detected by the model
classNames = ['Hardhat', 'Mask', 'NO-Hardhat', 'NO-Mask', 'NO-Safety Vest', 'Person', 'Safety Cone',
              'Safety Vest', 'machinery', 'vehicle']


prediction_man = model(img_man)
prediction_group = model(img_group)
```

# Print Results

```python
 # iterate over each box
  
# show the image
# Display the image
msaai.drawPredictions(prediction_man, img_man, classNames)
plt.imshow(img_man)
plt.show()

msaai.drawPredictions(prediction_group, img_group, classNames)
plt.imshow(img_group)
plt.show()
```
